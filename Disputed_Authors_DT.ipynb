```python
# We are focusing on Hamilton and Madison for authors
# Removing rest of the authors.
data_updated=data[(data['author']!='Jay') & (data['author']!='HM') &(data['author']!='dispt')]
data_updated.head()
```

```python
# No. of Essay's written by Hamilton and Madison
data_updated['author'].value_counts()
```

```python
# Creating X and y for independent and dependent features.
# Feature Variables in X
X=data_updated.drop(['filename','author'],axis=1)
#X=X.loc[:,~X.columns.str.startswith('author_')]
#Response Variable in y
y=data_updated['author']
```

```python
# Train test Split
# Carrying 75 25 split.
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=42)
print(X_train.shape,X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# X_train 
X_train.head()
```

```python
# X_test
X_test.head()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

# Instantiating Decision Tree Classifier
dt=DecisionTreeClassifier()

# Fitting the decision tree on X_train and y_train.
dt.fit(X_train,y_train)
```

```python
# Importing required packages for visualization
from IPython.display import Image  
from six import StringIO  
from sklearn.tree import export_graphviz
import pydotplus, graphviz
```

#### Without Hyperparameter Tuning

```python
# Plotting tree
dot_data = StringIO()  
export_graphviz(dt, out_file=dot_data, filled=True,rounded=True,
                feature_names=X.columns,
                class_names=['Hamilton','Madison']) 
                 
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
# Tree Interpretation

# We begin with a total of 46 essays out of which 35 belong to Hamilton  and 11 belong to Madison i.e., our Root Node. 
# We have the gini index as 0.364 that specifies the degree of randomly selecting a data point being classified incorrectly.
# If the Gini Index is 0 we get complete homogeneity where all the data points belong to 1 class itself.

# The split is done on word 'upon' becuase it has high Information gain amongst all other words.
# 1st Question we ask is: If the percentage of upon in an essay is less than or equal to 0.015%?

# Right Path
# We reach the leaf node where 34 essays belong to Hamilton if the presence of word upon in an essay is greater than 0.015%

# Left Path
# If the answer to that is Yes we go to the left and if it's No we go to the right.
# If we take the left path we get a total of 12 essays out of which one belongs to Hamilton and Madison if the percentage of not in an essay is less than 0.064%
# Then we further split that node and reach the leaf node and on the left we have ginidex as 0 for Hamilton where he has only 1 essay if the percentage of not is less than 0.064%
# If the percentage is greater than 0.064% we all the 11 essays from Madison with gini index being 0.
```

```python
# Evaluating Model performamce
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)
```

```python
# Importing necessary libraries from sklearn
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Model Accuracy on Train Set
print(accuracy_score(y_train, y_train_pred))
confusion_matrix(y_train, y_train_pred)
```

```python
# Model Accuracy of Test set
print(accuracy_score(y_test, y_test_pred))
confusion_matrix(y_test, y_test_pred)
```

```python
# Creating a Function to create decision tree
def get_dt_graph(dt_classifier):
    dot_data = StringIO()
    export_graphviz(dt_classifier, out_file=dot_data, filled=True,rounded=True,
                    feature_names=X.columns,
                    class_names=['Hamilton','Madison'])
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    return graph
```

```python
# Creating a Function for Model classification
def evaluate_model(dt_classifier):
    print("Train Accuracy :", accuracy_score(y_train, dt_classifier.predict(X_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, dt_classifier.predict(X_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, dt_classifier.predict(X_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, dt_classifier.predict(X_test)))
```

```python
# Decision Tree Classifier
dt_default = DecisionTreeClassifier(random_state=42)
dt_default.fit(X_train, y_train)
```

```python
evaluate_model(dt_default)
```

```python
# We are getting a Train Accuracy of 100% and Test Accuracy of 95% that shows we have Overfitting scenario.
# In order to control that we need to do hyperparameter Tuning
```

#### With Hyper parameter Tuning

```python
dt = DecisionTreeClassifier(random_state=42)
```

```python
# Importing Grid Search CV from sklearn
from sklearn.model_selection import GridSearchCV
```

```python
# Creating parameter grid based on the results of random search 
# Here we have specified the max_depth,minimum samples at leaf node and criteron i.e., gini indx or entropy
params = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ["gini", "entropy"]
}
```

```python
# Instantiate the grid search model
# In the model we are carrying a 4 fold cross validation.
# This means that there will be a total of 50 possibilites i.e., 5 entries in max depth, 5 minimum samples at leaf node
# There are two criterion. This implies 5*5*2=50 posibilities.
# No. of fits will be 50*4 i.e. 200
grid_search = GridSearchCV(estimator=dt, 
                           param_grid=params, 
                           cv=4, n_jobs=-1, verbose=1, scoring = "accuracy")
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
# Creating a dataframe to 
score_df = pd.DataFrame(grid_search.cv_results_)
score_df.head()
```

```python
# Using grid_search.best_estimator_ we get the best D.T. Classifier
grid_search.best_estimator_
```

```python
dt_best = grid_search.best_estimator_
```

```python
# Finding the Train and test Accuracy of the model 
evaluate_model(dt_best)
# Train Accuracy comes out to be 90.4%
# Test Accuracy comes out to be 86.3%
```

```python
# After we have done the hyperparameter tuning we get an accuracy of 97% on Train data and 95% on Test data.
```

```python
# Classification Report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, dt_best.predict(X_test)))
```
